```python
import rasterio
import matplotlib.pyplot as plt

file_path = '/data/fast1/nsidc0766/downloaded_files/GL_vel_mosaic_s1cycle_18Jul16_29Jul16_vv_v02.0.tif'
# Open the GeoTIFF file
with rasterio.open(file_path) as src:
    # Read the first band
    band1 = src.read(1)
    print(src)
    
    # Get the metadata
    print(f"Metadata: {src.meta}")
    
    # Plot the data
    plt.figure(figsize=(10, 6))
    plt.imshow(band1, cmap='viridis',vmin=0,vmax=1000)
    plt.colorbar(label='Digital Elevation')
    plt.title('GeoTIFF Data Plot')
    plt.xlabel('Longitude index')
    plt.ylabel('Latitude index')
    plt.ylim([8400,8000])
    plt.xlim(2300,2800)
    plt.show()
```
